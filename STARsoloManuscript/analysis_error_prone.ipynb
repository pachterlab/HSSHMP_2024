```python
results_df_template = pd.DataFrame({
    "Mismatches": [],
    "Deletions": [],
    "Insertions": [],
    "Program": [],
    "Run mode": [],
    "Median ρ*": [],
    "Median r": [],
    "RMSE": [],
    "FPR": [],
    "FNR": [],
})

def get_results(data_dict):

    results = data_dict.copy()
    results.pop('spearman_corrs', None)
    results.pop('spearman_corrs_zeroes_included', None)
    results.pop('errors', None)
    results.pop('false_positives', None)
    results.pop('false_negatives', None)
    results.pop('num_genes', None)
    results.pop('total_count', None)
    results.pop('num_barcodes_in_intersection', None)
    results.pop('num_barcodes_in_simulation', None)
    results.pop('num_barcodes_in_program', None)
    results.pop('num_genes_in_intersection', None)
    results.pop('num_genes_in_simulation', None)
    results.pop('num_genes_in_program', None)
    rmse = results['rmse']
    fnr = results['fnr']
    fpr = results['fpr']
    med_rho = results['median_spearman_corrs']
    med_rho_0 = results['median_pearson_corrs_zeroes_included']
    return [med_rho, med_rho_0, rmse, fpr, fnr]

results_df = results_df_template.copy()
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "kallisto", "standard (no D-list)"] +
    get_results(results_sim_vs_kallisto_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "kallisto", "standard (D-list)"] +
    get_results(results_sim_vs_offlist_kallisto_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "kallisto", "nac (no D-list)"] +
    get_results(results_nac_sim_vs_kallisto_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "kallisto", "nac (D-list)"] +
    get_results(results_nac_sim_vs_offlist_kallisto_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "alevin-fry", "splici align"] +
    get_results(results_sim_vs_af_splici_rad_crlike_sa_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "alevin-fry", "splici sketch"] +
    get_results(results_sim_vs_af_splici_sketch_crlike_sa_1_1_1)
)
results_df.loc[len(results_df)] = (
    ["1", "1", "1", "STARsolo", ""] +
    get_results(results_sim_vs_star_1_1_1)
)

results_df.loc[len(results_df)] = (
    ["4", "0", "0", "kallisto", "standard (no D-list)"] +
    get_results(results_sim_vs_kallisto_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "kallisto", "standard (D-list)"] +
    get_results(results_sim_vs_offlist_kallisto_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "kallisto", "nac (no D-list)"] +
    get_results(results_nac_sim_vs_kallisto_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "kallisto", "nac (D-list)"] +
    get_results(results_nac_sim_vs_offlist_kallisto_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "alevin-fry", "splici align"] +
    get_results(results_sim_vs_af_splici_rad_crlike_sa_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "alevin-fry", "splici sketch"] +
    get_results(results_sim_vs_af_splici_sketch_crlike_sa_4_0_0)
)
results_df.loc[len(results_df)] = (
    ["4", "0", "0", "STARsolo", ""] +
    get_results(results_sim_vs_star_4_0_0)
)


results_df.loc[len(results_df)] = (
    ["0", "2", "2", "kallisto", "standard (no D-list)"] +
    get_results(results_sim_vs_kallisto_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "kallisto", "standard (D-list)"] +
    get_results(results_sim_vs_offlist_kallisto_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "kallisto", "nac (no D-list)"] +
    get_results(results_nac_sim_vs_kallisto_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "kallisto", "nac (D-list)"] +
    get_results(results_nac_sim_vs_offlist_kallisto_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "alevin-fry", "splici align"] +
    get_results(results_sim_vs_af_splici_rad_crlike_sa_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "alevin-fry", "splici sketch"] +
    get_results(results_sim_vs_af_splici_sketch_crlike_sa_0_2_2)
)
results_df.loc[len(results_df)] = (
    ["0", "2", "2", "STARsolo", ""] +
    get_results(results_sim_vs_star_0_2_2)
)

results_df
```
